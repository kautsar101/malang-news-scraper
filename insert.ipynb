```python
from supabase import create_client
from dotenv import load_dotenv
import os

load_dotenv()

supabase = create_client(
    os.getenv("SUPABASE_URL"),
    os.getenv("SUPABASE_KEY")
)

data = {
    "title": "Test Article",
    "author": "Kautsar",
    "published_date": "2026-06-18",
    "content": "Ini test insert dari Python",
    "url": "https://test.com",
    "source": "test"
}

result = (
    supabase
    .table("news_articles")
    .insert(data)
    .execute()
)

print(result)
```
